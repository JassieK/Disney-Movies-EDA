```python
# Reading our Data 
df = pd.read_csv(r'D:\Data Analytics\Disney Movies\disney_movies.csv')
```

```python
df.head()
```

```python
# Checking if there any missing values in our data 
for col in df.columns :
    pct_miss = np.mean (df[col].isnull())
    print ('{}-{}%' .format(col,pct_miss))
        
```

```python
# Dropping the rows that contain null values 
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
# There are rows in our mpaa_rating coloumns which says 'Not Rated' replacing those values by NULL
df['mpaa_rating'] = df['mpaa_rating'].replace(['Not Rated'] , None )
```

```python
df.isnull().sum()
```

```python
# Now Dropping the NULL values 
df = df.dropna()
```

```python
pd.set_option('display.max_rows',None)
```

```python
# Checking the data types
df.dtypes
```

```python
# Making a new coloumn for release year 
df['Year'] = df['release_date'].astype(str).str[:4]
df
```

```python
df['Year'] =df['Year'].astype('int64')
df.info()
```

```python
X = []
for i in range (510):
    if df['Year'].iloc[i]  > 2010 :
        X.append('2010 - 2020')
        
    elif df['Year'].iloc[i] <= 2010 and df['Year'].iloc[i] > 2000 :
        X.append('2000 - 2010')
        
    elif df['Year'].iloc[i] <= 2000 and df['Year'].iloc[i] > 1990 :
        X.append('1990 - 2000')
        
    elif df['Year'].iloc[i] <= 1990 and df['Year'].iloc[i] > 1980 :
        X.append('1980 - 1990')
        
    elif df['Year'].iloc[i] <= 1980 and df['Year'].iloc[i] > 1970 :
        X.append('1970 - 1980')
        
    elif df['Year'].iloc[i] <= 1970 and df['Year'].iloc[i] > 1960 :
        X.append('1960 - 1970')
        
    elif df['Year'].iloc[i] <= 1960 and df['Year'].iloc[i]> 1950 :
        X.append('1950 - 1960')
        
    else:
        X.append('< 1950')
df['decade'] = X
       
 
```

```python
decade_lis =  df['decade'].unique()
fig,axes = plt.subplots(4,2,figsize=(20,20))
axes = axes.flatten()
for i in range(len(decade_lis)):
    ax = sns.countplot(y='genre',data=df[df['decade']==decade_lis[i]],ax=axes[i],order=df[df['decade']==decade_lis[i]]['genre'].value_counts().index)
    ax.set_title(decade_lis[i])
    fig.show()
plt.show()
```

```python
# Sorting the data according to the highest gross amount
df = df.sort_values( by = ['total_gross'] , inplace = False , ascending = False)
```

```python
#as correlation only works on int values , lets assign other objects int values
df_num = df 
for col_name in df_num.columns:
    if (df_num[col_name ].dtype == 'object') :
        df_num[col_name ] = df_num[col_name ].astype('category')
        df_num[col_name ] = df_num[col_name ].cat.codes
df_num       
```

```python
# lets find the correlation on total data
df_num.corr()
```

```python
# As we can see gross is highly correlated with release date
plt.scatter (x = df ['Year'] , y = df['total_gross'])
plt.title('Gross over the Years')
plt.xlabel('Year')
plt.ylabel('Gross')
plt.show()
```
